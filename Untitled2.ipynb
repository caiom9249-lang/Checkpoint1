```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [20, 40, 25, 30]

plt.plot(x, y)
plt.title("gráfico simples")
plt.show()
```

explicação:

Esse código serve para criar um gráfico simples.

Ele começa importando uma ferramenta (matplotlib) que permite desenhar gráficos. Depois, define duas listas:

x: posições no eixo horizontal
y: valores que vão aparecer no eixo vertical

Cada número de x é combinado com um número de y, formando pontos no gráfico.

Em seguida, plt.plot(x, y) liga esses pontos com uma linha, formando o gráfico.
plt.title("gráfico simples") coloca um título.
E plt.show() mostra o gráfico na tela.
